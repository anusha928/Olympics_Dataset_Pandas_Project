```python
coffee["revenue"] = coffee["Units Sold"] * coffee["price"]
```

```python
# shift
coffee["yesterdayRevenue"] = coffee["revenue"].shift(2)
```

```python
coffee["%_change"] = ((coffee["revenue"] - coffee["yesterdayRevenue"])/ coffee["yesterdayRevenue"])* 100
```
