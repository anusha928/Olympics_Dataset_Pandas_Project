```python
# manipulating date
```

```python
bios_new['birth_date'] = pd.to_datetime(bios_new['born_date'],format ="%Y-%m-%d" )
```

```python
bios_new.info()
```

```python
bios_new['born_year'] = bios_new['birth_date'].dt.year
```

```python
# determining the age
```

```python
from datetime import date

current_date = date.today()
current_year = current_date.year
print(current_year)
bios_new['age'] = current_year - bios_new['born_year']
```

```python
bios_new[['name','age']].query('age > 180')
```

```python
# saving the change in csv
```

```python
bios_new.to_csv('./data/bios_new.csv', index = False)
```

```python
# lamda function
```

```python
# determining age using lamda 

bios_new = bios_new.assign(age_lamda = lambda x:(current_year - x['born_year'] ))
```

```python
bios_new['height_category'] = bios_new['height_cm'].apply(lambda x: 'Short' if x < 180 
                          else 'Medium' if x < 210
                          else 'Tall'  )
```

```python
bios_new
```

```python
# using function
def categorize_athlete(row):
    if row['height_cm'] < 180 and row['weight_kg'] < 80:
        return 'LightWeight'
    elif row['height_cm'] < 200 and row['weight_kg'] <= 100:
        return "AverageWeight"
    else: 
        return "HeavyWeight"
    
bios_new['category'] = bios_new.apply(categorize_athlete, axis= 1)
```
